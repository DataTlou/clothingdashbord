```python
import pandas as pd
```

```python
df = pd.read_csv('MANUFACTURING_ALL.csv')
df
```

```python
df['NAME.1'].unique()
```

```python
df.dtypes
```

```python
utili = df[df['NAME.1'] == 'Utilisation']
under = df[df['NAME.1'] == 'Under-utilisation total']
raw = df[df['NAME.1'] == 'Reason for under-utilisation: Raw materials']
skilled = df[df['NAME.1'] == 'Reason for under-utilisation: Skilled labour']
semi = df[df['NAME.1'] == 'Reason for under-utilisation: Semi- and unskilled labour']
demand = df[df['NAME.1'] == 'Reason for under-utilisation: Insufficient demand']
other = df[df['NAME.1'] == 'Reason for under-utilisation: Other reasons']
```

```python
utili
```

```python
uu = utili.merge(under, on=['NAME', 'QUARTER', 'YEAR'], how= 'left')
uu
```

```python
uur = uu.merge(raw, on=['NAME', 'QUARTER', 'YEAR'], how='left')
uur
```

```python
uurs = uur.merge(skilled, how='left', on=['NAME', 'QUARTER', 'YEAR'])
uurs
```

```python
uurss = uurs.merge(semi, how='left', on=['NAME', 'QUARTER', 'YEAR'])
uurss
```

```python
uurssd = uurss.merge(demand, how='left', on=['NAME', 'QUARTER', 'YEAR'])
uurssd
```

```python
uurssdo = uurssd.merge(other,  how='left', on=['NAME', 'QUARTER', 'YEAR'])
uurssdo
```

```python
merged = uurssdo.copy()
merged
```

```python
merged.columns
```
